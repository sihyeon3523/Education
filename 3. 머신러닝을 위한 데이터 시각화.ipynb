# 라인 플롯

```python
x = np.linspace(0, 10, 100)
# .plot = 라인 플롯을 그리는 라이브러리 
plt.plot(x, np.sin(x), "-", linewidth=1, label="sin(x)") # "-" solid style
plt.plot(x, np.cos(x), "--", linewidth=1, label="cos(x)") # "--" dotted style 
plt.xlim(0, 10)
plt.ylim(-1, 1); # 경계 영역 지정 
plt.title("Sine & Cosine Curve")
plt.xlabel("x")
plt.legend(); # 범례 지정 
```

```python
with plt.style.context("ggplot"): # plot 스타일의 일시적 변경 (이 셀에서만 적용)
    plt.plot(x, np.sin(x), '-.', linewidth=1, label="sin(x)")
    plt.plot(x, np.cos(x), ':', linewidth=1, label="cos(x)")
    plt.xlim(0, 10)
    plt.ylim(-1, 1)
    plt.title("Sine & Cosine Curve")
    plt.xlabel("x")
    plt.legend();
```

### 객체지향 인터페이스 Figure(면에 대한 정보, 캔버스)와 Axes(축에 대한 정보)객체의 사용 

```python
fig = plt.figure() # 면
ax = plt.axes() # 축 

x = np.linspace(0, 10, 100)

# plt 대신 ax 사용. 거기에서 세터 (set)를 사용해서 값 설정 
ax.plot(x, np.sin(x), "-", linewidth=1, label="sin(x)")
ax.plot(x, np.cos(x), "--", linewidth=1, label="cos(x)")
ax.set_xlim([0, 10])
ax.set_ylim([-1, 1])
ax.set_title("Sine & Cosine Curve")
ax.set_xlabel("x")
ax.legend();
```

```python
fig, ax = plt.subplots(2, 1, sharex=True) # 도면 분리, 두개의 그래프 따로 그리기 

ax[0].plot(x, np.sin(x), 'b-', linewidth=0.5, label="sin(x)") # 'b-' blue, 실선 
ax[0].set_title("Sine & Cosine Curve")
ax[0].set(xlim=[0, 10], ylim=[-1, 1])
ax[0].legend()

ax[1].plot(x, np.cos(x), 'g-', linewidth=0.5, label="cos(x)") # 'g-', green 실선 
ax[1].set(xlim=[0, 10], ylim=[-1, 1], xlabel="x") # set 메소드에 속성값 파라미터로 지정
ax[1].legend();
```

```python
from scipy import stats

t_dist = stats.t(df=20)
t_values = np.linspace(-5, 5, 1000)

t_alpha = -1.729
pdf_values = t_dist.pdf(t_values)
fill_color = (0, 0, 0, 0.1)
line_color = (0, 0, 0, 0.5)
ax = plt.axes()
ax.plot(t_values, pdf_values)
ax.fill_between(t_values, pdf_values, where=t_values <= t_alpha, color=fill_color)
pd = t_dist.pdf(t_alpha)
ax.plot([t_alpha, t_alpha], [0, pd], color=line_color);
ax.set(xlabel='t value', ylabel='probability for t value')
ax.set_title('PDF for t distribution with df=20');
```

```python
from scipy import stats

mu = 0
std = 1
rv = stats.norm(mu, std)
x = np.linspace(-5, 5, 1000)
pdf_values = rv.pdf(x)

t_alpha = -1.729

fill_color = (0, 0, 0, 0.1)
line_color = (0, 0, 0, 0.5)

ax = plt.axes()
ax.plot(x, pdf_values)

ax.fill_between(x, pdf_values, where=x <= t_alpha, 
                color=fill_color)
pd = rv.pdf(t_alpha)
ax.plot([t_alpha, t_alpha], [0, pd], color=line_color);
ax.set(xlabel='x value', ylabel='probability for x value')
ax.set_title('PDF for gaussian normal distribution');
```

## iris 데이터 활용

```python
from seaborn import load_dataset
iris = load_dataset("iris") # 분류 문제에서 자주 활용되는 iris 데이터 
iris.info()
```

```python
iris.head() # species 출력, 나머지는 input
```

```python
import pandas as pd

iris["species_cat"] = pd.Categorical(iris.species) # Categorical함수를 이용해 columns를 범주형 데이터로 변환
iris.info()
```

```python
plt.scatter(iris.sepal_length, iris.sepal_width,
            c=iris.species_cat.cat.codes, alpha=0.2, cmap="viridis") 
# x,y,cat객체는 수치형 데이터를 가지고 있어 컬러 지정 가능, 컬러맵을 이용해 컬러 결정
# 품종을 컬러로 표현하여 경향에 대한 직관을 얻을 수 있음 
plt.title("iris")
plt.xlim(iris.sepal_length.min(), iris.sepal_length.max()) # 경계 값 지정 
plt.ylim(iris.sepal_width.min(), iris.sepal_width.max())
plt.xlabel(iris.columns[0]) # 문자열 정보 넣어주기 
plt.ylabel(iris.columns[1])
plt.colorbar(ticks=range(3), label="species")
```

```python
species = iris.species.unique() # unique() 함수를 이용해 3가지 품종에 대한 고유한 값 얻기 가능 
species
```

```python
for sp in species:
    x = iris.loc[iris["species"] == sp, "sepal_length"]
    y = iris.loc[iris["species"] == sp, "sepal_width"]
    plt.scatter(x, y, label=sp, alpha=0.2)
plt.title("iris")
plt.xlim(iris.sepal_length.min(), iris.sepal_length.max())
plt.ylim(iris.sepal_width.min(), iris.sepal_width.max())
plt.xlabel(iris.columns[0])
plt.ylabel(iris.columns[1])
plt.legend(); # legend()를 통해 label 정보 출력 
```

# 히스토그램

데이터의 분포 형상을 식별하는 히스토그램 

```python
np.random.seed(0)

mu = 72 # 평균값
sigma = 8 # 표준편차 
x = np.random.normal(mu, sigma, 10000) # 평균값=72,표준편차=8,10000건의 데이터 구성 

plt.hist(x, rwidth=0.9, bins=16); 
# bin =16 전체 계급 구간(x축이 나눠진 구간)이 16개,rwidth = 히스토그램 굵기, y=빈도수
```

```python
tips = load_dataset("tips") # seaborn 에 tips 에는 bill,tip,성별, 흡연 등의 정보가 들어있음
tips.info()
```

```python
plt.hist(x = tips.tip, rwidth=0.9, bins = 20); # 계급구간(20개) 얼마나 많은 빈도수(ylabel)가 있는가? 
plt.title("tips")
plt.xlabel("tip")
plt.ylabel("frequency");
```

```python
import scipy as sp # scipy=통계와 관련된 함수들이 내장

bins = 20
plt.hist(x = tips.tip, rwidth=0.9, bins = bins, density=True);
plt.title("tips")
plt.xlabel("tip")
plt.ylabel("density") # 정규분포의 근사치 얻기 = 확률 값이 나온다

mu, sigma = tips.tip.mean(), tips.tip.std() #평균, 표준편차
tip_min = tips.tip.min()
tip_max = tips.tip.max()
x = np.linspace(tip_min, tip_max, bins) 
y = sp.stats.norm(mu, sigma).pdf(x) # 확률밀도 함수(pdf)를 활용해 y값 얻기 
plt.plot(x, y);
```

# 박스 플롯

    박스플롯과 사분위수
    박스의 중간 구분선이 전체 데이터의 50%(Q2), 박스 하단은 25%(Q1), 박스 상단은 75%(Q3)를 나타냄 
    IQR: Inter Quantile Range = Q3 - Q1
    정상적인 데이터(펜스)까지 수염을 그림 
    (Q1-1.5)*IQR = Lower outer fence, (Q3+1.5)*IQR = Upper outer fence
    fence 밖의 점들을 이상치라고 한다. 
    박스플롯은 이상치 발견의 직관적인 도구 

```python
tips.loc[tips.sex == "Male", "tip"].describe().round(1) # describe() 요약통계량 
```

```python
tips.loc[tips.sex == "Female", "tip"].describe().round(1)
```

```python
labels = []
tip_list = []

for label, df_per_sex in tips.groupby("sex"):
    labels.append(label)
    tip_list.append(df_per_sex["tip"].tolist())

plt.boxplot(tip_list, labels=labels) 
# tip_list:tip의 정보, labels:male,female 
# 주로 x축에는 범주형 데이터 사용 
plt.title("tips")
plt.xlabel("sex")
plt.ylabel("tip");
```

# 이미지 플롯

plt.imshow() 를 이용한 이미지 플롯

```python
from sklearn.datasets import load_digits # sklearn.datasets의 load_digits 함수로 손글씨 데이터 로딩 

digits = load_digits() 
digits.keys() # bunch 라는 객체는 keys 함수 제공
```

```python
digits.images[0] 
# 이미지 정보는 픽셀 정보(8*8=64)로 구성되어 있음 
# 손글씨 이미지는 밝기 정보를 이용해 표현해 놓음 
# 값이 0인 경우에는 메모리만 차지하는 데이터로 희소행렬을 이용해 변환해야 한다. 
```

```python
fig, ax = plt.subplots(10, 10, figsize=(5, 5)) # 10행 10열 100개의 그림 그리기 
for i, ax_i in enumerate(ax.flat): 
    # enumerate 함수를 이용해 figure 객체와 axes 얻기, ax.flat 시키기(2차원 구조 -> 1차원 구조)
    ax_i.imshow(digits.images[i], cmap="binary") # binary: 2진 형태=흑백
    ax_i.set(xticks=[], yticks=[]) # x,y 축 정보를 기술하지 않고 공간만 구성 
```

    필기 인식이 얼만큼 잘 되는지 라벨을 주어 그래픽 겨로가를 만들어 내는 데에 이미지 플롯이 유용함
    이미지 플롯은 데이터를 읽어내 어떻게 이미지로 그려낼지에 관한 것이다. 

# 주요 정리

1. matplotlib 라이브러리를 이용해 시각화 작업을 할 경우, figure와 axes 객체와 같은 객체 지향 인터페이스를 사용하거나, 활성화된 그림이나 축을 사용하는 매트랩 스타일의 pylab 인터페이스를 사용할 수 있다
2. 라인 플롯은 함수를 시각화하기 위한 플롯으로, plot() 함수를 사용하며, 선의 색상, 스타일과 같은 속성을 데이터에 매핑해 사용한다.
3. 스캐터 플롯은 데이터의 산포도에 대한 직관을 얻을 수 있는 플롯으로, scatter() 함수를 사용하며, 각 점의 크기, 색상과 같은 속성을 데이터에 매핑해 사용한다. 
4. 히스토그램은 계급 구간별 빈도와 밀도를 쉽게 이해할 수 있는 플롯으로 hist()함수를 사용하며, scipy.stats 패키지의 norm.pdf()함수를 이용해 근사값을 같이 표시할 수 있다
5. 박스 플롯은 사분위수와 이상치의 존재 여부를 쉽게 파악할 수 있는 플롯으로 boxplot() 함수를 사용하며, 이미지 플롯은 이미지 데이터를 개체에 시각화 하는 Imshow() 함수를 사용한다.
